```python
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
majorLocator = MultipleLocator(1)
majorFormatter = FormatStrFormatter('%d')
minorLocator = MultipleLocator(5)

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
xscale = ax.set_xscale('log')

__=ax.plot(1+z, 2.5*mah(z, 0.1, 0.69), label=r'Overall', color='k')
__=ax.plot(1+z, 2.5*mah(z, -0.04, 0.54), label=r'Type 1', color='cyan')
__=ax.plot(1+z, 2.5*mah(z, -0.9, 0.35), '--', label=r'Type 2', color='green')
__=ax.plot(1+z, 2.5*mah(z, 0.62, 0.88), '-.', label=r'Type 3', color='red')
__=ax.plot(1+z, 2.5*mah(z, 1.42, 1.39), '-.', label=r'Type 4', color='purple')

ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_major_formatter(majorFormatter)

# for the minor ticks, use no labels; default NullFormatter
ax.xaxis.set_minor_locator(minorLocator)
xlim = ax.set_xlim(1, 7)
ylim = ax.set_ylim(0, 3)
legend = ax.legend()
xlabel = ax.set_xlabel(r'1 + z', size=15)
ylabel = ax.set_ylabel(r'${\rm M(z) / M_0}$', size=15)

# To set the ticksize using axis notation
xticksize, yticksize = 13, 13
xticks1 = ax.xaxis.get_major_ticks()
yticks1 = ax.yaxis.get_major_ticks()
for xtick, ytick in zip(xticks1, yticks1):
    xtick.label.set_fontsize(xticksize) 
    ytick.label.set_fontsize(yticksize) 

fig.savefig('fig2_mcbride09_comparison.pdf', 
            bbox_extra_artists=[xlabel, ylabel], bbox_inches='tight')
```
